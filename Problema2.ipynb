```python
import pandas as pd

df = pd.read_csv('data/winemag-data-130k-v2.csv', index_col=0)
print("Información del DataFrame:")
print(df.info())
print("\nPrimeras 5 filas del DataFrame:")
print(df.head())

# Renombrar columnas
df.rename(columns={'country': 'pais', 'points': 'puntuacion', 'price': 'precio', 'variety': 'variedad'}, inplace=True)

# Crear nuevas columnas
# 1. Calificación del vino (basado en la puntuación)
def calificacion_vino(puntuacion):
    if puntuacion >= 95:
        return 'Excelente'
    elif puntuacion >= 90:
        return 'Muy bueno'
    elif puntuacion >= 85:
        return 'Bueno'
    else:
        return 'Regular'

df['calificacion'] = df['puntuacion'].apply(calificacion_vino)

# 2. Rango de precios
def rango_precio(precio):
    if pd.isnull(precio):
        return 'Desconocido'
    elif precio < 20:
        return 'Barato'
    elif precio < 50:
        return 'Moderado'
    else:
        return 'Caro'

df['rango_precio'] = df['precio'].apply(rango_precio)

# 3. Continente del país
def continente(pais):
    if pais in ['France', 'Italy', 'Spain', 'Portugal', 'Germany', 'Austria', 'Greece', 'Hungary']:
        return 'Europa'
    elif pais in ['US', 'Canada', 'Mexico', 'Argentina', 'Chile', 'Brazil', 'Uruguay']:
        return 'América'
    elif pais in ['Australia', 'New Zealand']:
        return 'Oceanía'
    else:
        return 'Otros'

df['continente'] = df['pais'].apply(continente)

#Reporte 1: Prom. puntuación y precio x país
reporte1 = df.groupby('pais').agg({'puntuacion': 'mean', 'precio': 'mean'}).sort_values(by='puntuacion', ascending=False)
print("\nPromedio de puntuación y precio por país:")
print(reporte1.head())

#Reporte 2: Cant. de vinos x calificación
reporte2 = df['calificacion'].value_counts().sort_index()
print("\nCantidad de vinos por calificación:")
print(reporte2)

#Reporte 3: Vinos mejor puntuados x continente
reporte3 = df.groupby('continente').apply(lambda x: x[x['puntuacion'] == x['puntuacion'].max()])
print("\nVinos mejor puntuados por continente:")
print(reporte3[['pais', 'variedad', 'puntuacion']])

#Reporte 4: % de precio y cantidad de vinos x rango de precios
reporte4 = df.groupby('rango_precio').agg({'precio': 'mean', 'variedad': 'count'}).rename(columns={'variedad': 'cantidad_vinos'})
print("\nPromedio de precio y cantidad de vinos por rango de precios:")
print(reporte4)

# Guardar el reporte 1 en Excel
reporte1.to_excel('reporte1.xlsx')
print("\nReporte 1 guardado como 'reporte1.xlsx'.")

```
